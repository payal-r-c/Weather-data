```python
num_df = df1[{'Date/Time','Temp_C','Dew Point Temp_C','Rel Hum_%','Wind Speed_km/h','Visibility_km','Press_kPa','Weather'}]
```

```python
corrmat = num_df.corr()
f,ax = plt.subplots(figsize = (12,7))
sns.heatmap(corrmat,annot = True, fmt = '.2f', annot_kws = {'size' : 10}, vmax = .8 , square = True);
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(y = 'Wind Speed_km/h',x = 'Press_kPa', data = df1)
plt.show()
```

```python
df1.drop(df1[df1['Wind Speed_km/h'] >5000].index,inplace = True)
```

```python
plt.figure(figsize = (12,6))
sns.scatterplot(y = 'Wind Speed_km/h',x = 'Press_kPa', data = df1)
plt.show()
```

```python
d1 = pd.DataFrame(df1['Temp_C'].value_counts()).reset_index().rename(columns = {'index':'Temp_C','Temp_C':'Dew Point Temp_C'}).sort_values( 'Dew Point Temp_C',ascending = False)
d1.drop(d1[d1['Temp_C'] == 0].index,inplace = True)
d1 = d1[:10]
plt.figure(figsize = (10,5))
sns.barplot(x = 'Temp_C' , y = 'Dew Point Temp_C',data = d1 , order = d1.sort_values('Dew Point Temp_C',ascending = False).Temp_C)
```

```python
fig, axes = plt.subplots(1,2,figsize = (18,8))
grp_by_room = df1.groupby('Weather')
d1['Weather'] = grp_by_room.size()
sns.countplot(ax = axes[0],x = df1['Weather'])
sns.boxplot(ax = axes[1],x = df1['Weather'], y = df1['Wind Speed_km/h'])
plt.show()
```
